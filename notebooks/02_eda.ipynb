```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("laptops_featured.csv")
df.head()
```

```python
df.shape
df.describe()
df.isnull().sum()
```

```python
brand_price = df.groupby("Brand")["Price"].mean().sort_values(ascending=False)

plt.figure(figsize=(8,4))
brand_price.plot(kind="bar")
plt.title("Average Laptop Price by Brand")
plt.ylabel("Average Price")
plt.xlabel("Brand")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="RAM_GB", y="Price", data=df)
plt.title("Price Distribution by RAM Size")
plt.show()
```

```python
cpu_price = df.groupby("CPU_Brand")["Price"].mean()

plt.figure(figsize=(5,4))
cpu_price.plot(kind="bar", color="skyblue")
plt.title("Average Price by CPU Brand")
plt.ylabel("Average Price")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.boxplot(x="GPU_Type", y="Price", data=df)
plt.title("Price Comparison: Integrated vs Dedicated GPU")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Storage_GB", y="Price", hue="GPU_Type", data=df)
plt.title("Storage Size vs Price")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x="Screen_Size", y="Price", hue="Price_Category", data=df)
plt.title("Screen Size vs Laptop Price")
plt.show()
```
